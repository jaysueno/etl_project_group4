```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Big_Cities_Health_Data_Inventory.csv')
```

```python
df.head()
```

```python
df.head(10).T
```

```python
df['Indicator Category'].value_counts()
```

```python
df_clean = df[['Indicator Category','Indicator','Place','Value']]
```

```python
df_health = df_clean[df_clean['Indicator Category'] =='Life Expectancy and Death Rate (Overall)']
```

```python
df_health['Indicator'].value_counts()
```

```python
df_expectancy= df_health[df_health['Indicator']=='Life Expectancy at Birth (Years)']
```

```python
df_expectancy['Place'].value_counts()
```

```python
new = df_expectancy['Place'].str.split(',',n=1,expand=True)
```

```python
df_expectancy['city']=new[0]
```

```python
df_expectancy['state']=new[1]
```

```python
df_expectancy.drop(columns=['Place'],inplace = True)
```

```python
df_expectancy.head()
```

```python
df_expectancy['city'].value_counts()
```

```python
df_expectancy = df_expectancy.replace({'Portland (Multnomah County)':'Portland','San Diego County':'San Diego','Las Vegas (Clark County)':'Las Vegas','Fort Worth (Tarrant County)':'Fort Worth','Miami (Miami-Dade County)':'Miami','Atlanta (Fulton County)':'Atlanta'})
```

```python
df_expectancy = df_expectancy[['Value','city','state']]
```

```python
df_expectancy = df_expectancy.rename(columns={'Value':'life_expectancy'})
```

```python
df_expectancy = df_expectancy.reset_index(drop=True)
```

```python
df_expectancy.head()
```

```python
df_expectancy.to_csv('expectancy.csv')
```
